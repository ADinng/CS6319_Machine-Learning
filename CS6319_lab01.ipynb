```python
import pandas as pd

data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/parkinsons/telemonitoring/parkinsons_updrs.data")
data.head()
```

```python
data = data.drop(["subject#","total_UPDRS"], axis=1)
```

```python
data.describe()
```

```python
# creat training and testing data
from sklearn.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size = 0.2)
print(trainset)
print(testset)
```

```python
# separate the labels
traindata = trainset.drop("motor_UPDRS", axis=1)
trainlabs = trainset["motor_UPDRS"].copy()

testdata = testset.drop("motor_UPDRS", axis=1)
testlabs = testset["motor_UPDRS"].copy()
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
trainnumr = scaler.fit_transform(traindata)
testnumr = scaler.transform(testdata)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(trainnumr, trainlabs)
```

```python
test_predictions = linreg.predict(testnumr)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
test_mse = mean_squared_error(testlabs, test_predictions)
test_rmse = np.sqrt(test_mse)
print("The regressor's RMSE err on the test data : ", test_rmse)
```
